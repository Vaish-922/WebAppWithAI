```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data=[(2,6),(1,3),(3,9),(14,20),(12,16),(8,5),(10,6),(2,9),(5,7),(5,4),(9,12),(16,14),(18,11),(13,17),(9,11),(14,13),(11,5),(9,6),(13,1),(18,6)]


```

```python
def func(x1,x2,y1,y2):
  distance = abs(x2-x1)+abs(y2-y1)
  return distance

```

```python
def error(x1,x2,y1,y2):
  distance = (((x1-x2)**2)+((y1-y2)**2))**0.5
  return distance
```

```python
def find_min(ar):
  min_dis=min(ar)
  for l in range(len(ar)):
    if ar[l] == min_dis:
      return l
```

```python
er=[]

for i in range(1,6,1):

  centroid_old=[]
  centroid_new=[]
  cluster={}
  for j in range(i):
    centroid_new.append(data[j])
    cluster[j+1]=[]



  while(centroid_new!=centroid_old):
    centroid_old=centroid_new
    for l in data:
      ele_dist=[]
      for k in range(i):
        distance=func(centroid_old[k][0],l[0],centroid_old[k][1],l[1])
        ele_dist.append(distance)
      pos = find_min(ele_dist)
      cluster[pos+1].append(l)
    
    for c in range(i):
      centroid_new[c]=tuple(map(np.mean, zip(*cluster[c+1])))
      #centroid_new[c]= [sum(ele) / len(cluster[c+1]) for ele in zip(*cluster[c+1])]
      #for p in cluster[k+1]:
        #sum_x+=p[0]
        #sum_y+=p[1]
      #c=[sum_x/len(cluster[k+1]),sum_y/len(cluster[k+1])]
      #centroid_new[k]=[tuple([sum_x/len(cluster[k+1]),sum_y/len(cluster[k+1]])]
      


  c_x=[]
  c_y=[]
  for j in range(i):
    x=[]
    y=[]
    c_x.append(centroid_new[j][0])
    c_y.append(centroid_new[j][1])

    for k in cluster[j+1]:
      x.append(k[0])
      y.append(k[1])
    plt.scatter(x, y , label = j+1)
  plt.scatter(c_x, c_y, s = 80, color = 'k')
  plt.legend()
  plt.show()

  sum2=0
  for j in range(i):
    for k in cluster[j+1]:
      dist2=error(centroid_new[j][0],k[0],centroid_new[j][1],k[1])
      sum2=sum2+dist2
  
  er.append(sum2)
count=[1,2,3,4,5]
plt.plot(count,er)
plt.xlabel('No of Clusters')
plt.ylabel('Mean Error')
plt.title('Elbow graph')
plt.show()




```
